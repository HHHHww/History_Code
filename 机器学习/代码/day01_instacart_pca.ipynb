```python
import pandas as pd
```

```python
# 1、获取数据
aisles=pd.read_csv('F:/机器学习/Python3天快速入门机器学习项目资料/机器学习day1资料/02-代码/instacart/aisles.csv')
order_products=pd.read_csv('F:/机器学习/Python3天快速入门机器学习项目资料/机器学习day1资料/02-代码/instacart/order_products__prior.csv')
orders=pd.read_csv('F:/机器学习/Python3天快速入门机器学习项目资料/机器学习day1资料/02-代码/instacart/orders.csv')
products=pd.read_csv('F:/机器学习/Python3天快速入门机器学习项目资料/机器学习day1资料/02-代码/instacart/products.csv')
```

```python
aisles
```

```python
# 2、合并表
tab1=pd.merge(aisles,products,on=['aisle_id','aisle_id'])
```

```python
tab1
```

```python
tab2=pd.merge(tab1,order_products,on=['product_id','product_id'])
```

```python
tab2
```

```python
tab3=pd.merge(tab2,orders,on=['order_id','order_id'])
```

```python
tab3.head()
```

```python
# 3、找到user_id和aisle之间的关系
table=pd.crosstab(tab3['user_id'],tab3['aisle'])
```

```python
table
```

```python
data=table[:10000]
```

```python
# 4、PCA降维
from sklearn.decomposition import PCA
transfer=PCA(n_components=0.95)
data_new=transfer.fit_transform(data)
print(data_new.shape)
```

```python
data_new
```

```python
# 预估器流程
from sklearn.cluster import KMeans
```

```python
estimator=KMeans(n_clusters=3)
estimator.fit(data_new)
```

```python
y_predict=estimator.predict(data_new)
```

```python
y_predict[:300]
```

```python
# 模型评估-轮廓系数
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(data_new,y_predict)
```
